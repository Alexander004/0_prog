```python
import pandas as pd
import numpy as np
import scipy.stats

df = pd.read_csv('Data/MarvelComicCaracters.csv', sep = ';')
df
```

```python
df.shape[0] - df['appearances'].describe()[0]
```

```python
clone_df = df.dropna(subset=['appearances'])
```

```python
clone_df
```

```python
clone_df.groupby('align')['appearances'].describe()
```

```python
q1 = clone_df['appearances'].quantile(0.25)
q3 = clone_df['appearances'].quantile(0.75)
iqr = q3 - q1
print(iqr)
low = q1 - 1.5 * iqr
high = q3 + 1.5 * iqr
print(low, high)
```

```python
clone_df[clone_df['appearances'] > high].shape
```

```python
clone_df[clone_df['appearances'] > high]['align'].value_counts()
```

```python
mean = clone_df['appearances'].mean()
std = clone_df['appearances'].std()
std_outlier_threshold_bottom = mean - 2.5 * std
std_outlier_threshold_up = mean + 2.5 * std
print(std_outlier_threshold_bottom)
print(int(std_outlier_threshold_up))
```
